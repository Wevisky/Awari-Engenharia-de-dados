# Trabalhando com dicionários

```python
mw = {'CH4': 16.04, 'H2O': 18.02, 'O2':32.00, 'CO2': 44.01}
mw
```

```python
mw['C8H18'] = 114.23
mw
```

```python
mw['CH4']
```

```python
for tipos in mw.keys():
    print("A massa molar de {:<s} is {:<7.2f}".format(tipos, mw[tipos]))
```

## Lista sendo exibidas usando ordenação

```python
for tipos in sorted(mw):
    print(" {:<8s}  {:>7.2f}".format(tipos, mw[tipos]))
```

```python
for tipos in sorted(mw, key = mw.get):
    print(" {:<8s}  {:>7.2f}".format(tipos, mw[tipos]))
```

# Plotando gráficos

```python
%matplotlib inline

x = np.linspace(0,10)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,'b',x,z,'r')
plt.xlabel('Radianos');
plt.ylabel('Valor');
plt.title('Gráfico Demonstrativo')
plt.legend(['Seno','Coseno'])
plt.grid()
```

```python
plt.plot(y,z)
plt.axis('equal')
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title('Seno de (x)')

plt.subplot(2,1,2)
plt.plot(x,z)
plt.title('Coseno de (x)')
```
